## Importing the dataset

In the next step, we will use pandas read_csv command in order to import the dataset
Pay attention that the dataset is not stored in our pc, we are directly importing it by providing the url of the file.

* ### A note about the header
It is noteworthy that the original file does not have any header explaining what each column refers to. Hence, we are supposed to choose the header to be None, as if not pandas by automatically considers the first raw to be the header (a row that includes the names of all columns)



```python
url = url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
df = pd.read_csv(url,header=None)
```

## Having a look on the dataset 
In order to have a quick look on the dataset we can use the df.head(n) or df.tail(n) commands. Where DF.head(n) shows the first n rows of the dataset and df.tail(n) demosntrates the last n rows of it. if you do not provide any argument n, Pandas by default shows the first 5 rows

```python
df.head(10)
```

```python
df.tail()
```

## Adding the header
 As we previously saw the dataset did not have a header in the first raw, though the name of each column is available at the repository:
 https://archive.ics.uci.edu/ml/datasets/Automobile
 
In order to do se we will first create a lst called "headers" that includes all of the columns names 

```python
newHeader = ["symboling","normalized-losses","make","fuel-type","aspiration",
"num-of-doors","body-style","drive-wheels","engine-location","wheel-base","length",
"width","height","curb-weight","engine-type","num-of-cylinder","engine-size",
"fuel-system","bore","stroke","compression-ratio","horsepower","peak-rpm",
"city-mpg"," highway-mpg","price"]
newHeader
```

we will now use df.columns = headers to replace the existing numeric header with the one we created, and we will use df.head to have a look on the updated dataframe



```python
df.columns = newHeader
df.head()
```

We could also simply check the columns names of the updated dataframe through df.columns

```python
df.columns
```

we can now save the updated dataframe as a csv file

```python
file_path = "CarDataset.csv"
df.to_csv(file_path)
```
